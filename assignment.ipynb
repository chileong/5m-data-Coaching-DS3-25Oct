```python
import numpy as np
import pandas as pd
df = pd.read_csv('vgsales.csv')
df
```

Which 5 games have the highest global sales? What do they have in common?

```python
df.sort_values(by='Global_Sales',ascending=[False]).head(5)
```

Calculate total sales by genre. Which genre generates the most revenue globally?

```python
df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=[False])
```

Find the average sales per game for each platform. Which platforms have the highest average?

```python
df[df['Platform']=='GEN']
```

```python
df.groupby('Platform').agg(Global_Sales_sum=('Global_Sales','sum'),Global_Sales_mean=('Global_Sales','mean')).sort_values(by='Global_Sales_mean',ascending=[False])
```

Which publisher has released the most games? Do they also have the highest total sales?

```python
df.groupby('Publisher').agg(Name_count=('Name','nunique'),Sales=('Global_Sales','sum')).sort_values(by='Name_count',ascending=[False])
```

```python
df.groupby('Publisher').agg(Name_count=('Name','nunique'),Sales=('Global_Sales','sum')).sort_values(by='Sales',ascending=[False])
```

For games released after 2010, which genre dominated each region (NA, EU, JP)?

```python
df1=df[df['Year']>2010]
df1.groupby('Genre').agg(NA_Sales_sum=('NA_Sales','sum'),
                         EU_Sales_sum=('EU_Sales','sum'),
                         JP_Sales_sum=('JP_Sales','sum'),
                         Other_Sales_sum=('Other_Sales','sum'),
                         Global_Sales_sum=('Global_Sales','sum')
                         ).sort_values(by='Global_Sales_sum',ascending=[False])     
```

Calculate the percentage of global sales that each region contributes. Which region is most important?

```python
NA_Sales_sum = df['NA_Sales'].sum()
EU_Sales_sum = df['EU_Sales'].sum()
JP_Sales_sum = df['JP_Sales'].sum()
Other_Sales_sum = df['Other_Sales'].sum()
Global_Sales_sum = df['Global_Sales'].sum() 
print(f"NA_Sales_sum: {(NA_Sales_sum/Global_Sales_sum):.2f},EU_Sales_sum: {(EU_Sales_sum/Global_Sales_sum):.2f},JP_Sales_sum: {(JP_Sales_sum/Global_Sales_sum):.2f},Other_Sales_sum: {(Other_Sales_sum/Global_Sales_sum):.2f}")
```

Find games where Japanese sales exceed North American sales. What patterns do you notice about genre or
publisher?

```python
df1=df[df['JP_Sales']>df['NA_Sales']]
df1.groupby('Genre').agg(JP_Sales_sum=('JP_Sales','sum'),
                         Count=('Name','nunique')
                         ).sort_values(by='JP_Sales_sum',ascending=[False])
```

```python
df1.groupby('Publisher').agg(JP_Sales_sum=('JP_Sales','sum'),
                             Count=('Name','nunique')
                             ).sort_values(by='JP_Sales_sum',ascending=[False])
```
